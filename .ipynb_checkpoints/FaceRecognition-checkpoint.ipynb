# Facial Recognition with PCA and XGboost

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from time import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
import mglearn
```

```python
faces_image = np.load('olivetti_faces.npy')
faces_target = np.load('olivetti_faces_target.npy')

n_row = 64
n_col = 64

faces_data = faces_image.reshape(faces_image.shape[0], faces_image.shape[1] * faces_image.shape[2])

plt.imshow(faces_image[22],cmap='gray')

plt.show()
print('picture in grayscale')
n_samples = faces_image.shape[0]
X = faces_data
n_features = faces_data.shape[1]
```

```python
print("put lables to predict the id of all people in the dataset")
y = faces_target
n_classes = faces_target.shape[0]
```

```python
print("now wel create the dataset in terms of numbers by converting it into numpy arrays")
data= pd.DataFrame(X)
data['target'] = pd.Series(y, index=data.index)

data.to_csv("face_data.csv", index=False)
print("printing data.head()")
data.head()
```

### The dataset is transformed into numbers. now let us read and visualize the data
##### Functions show_orignal_images(pixels) takes an argument of pixels as a numpy array and reshapes it into a 64*64 pixel grayscale image

```python
def show_orignal_images(pixels):
	fig, axes = plt.subplots(6, 10, figsize=(11, 7),
	                         subplot_kw={'xticks':[], 'yticks':[]})
	for i, ax in enumerate(axes.flat):
	    ax.imshow(np.array(pixels)[i].reshape(64, 64), cmap='gray')	
	plt.show()    
```

```python
df = pd.read_csv("face_data.csv")
targets = df["target"]
pixels = df.drop(["target"],axis=1)
```

```python

print(np.array(pixels).shape)
```

```python
show_orignal_images(pixels)
print("Original Images")
```

### Split Dataset into train and test set. we have set the test_size to 0.2 implies that take 20% of data for testing and train the ML model with 80% of the dataset. 

######  We have set the random state to 20 in this model. Random state ensures that the splits that you generate are reproducible. Scikit-learn uses random permutations to generate the splits. The random state that you provide is used as a seed to the random number generator. This ensures that the random numbers are generated in the same order. 

```python
x_train, x_test, y_train, y_test = train_test_split(pixels, targets,test_size=0.2,random_state=20)
```

# Principal Component Analysis
### PCA is a machine learning model which is a part of unsupervised learning.Unsupervised learning is a type of machine learning algorithm used to draw inferences from datasets consisting of input data without labeled responses.

```python
pca = PCA(n_components=150).fit(x_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.show()
print('From the graph it can be seen that the curve nearly flattens after 120 components which indicates that out of all the 64*64 pixels (4096) features, 120 components cause the most variance')
```

## Eigenface
#### Eigenface is a set of all the eigen vectors of a given image. An eigen vector is a vector that retains its span when a linear transformation is applied on it and the factor by which it is increased or decreased is called its eigen value.

#### The set of eigen vectors helps in locating the features that cause most varience hence intensively used in facial recognition

### Computing the eigen vectors                                                                                                  
                                                                                                     
##### Let T be the matrix of preprocessed training examples. The covariance matrix can then be computed as S = T T'   and the eigenvector decomposition of S is given by


                
##### However T T' is a large matrix, and if instead we take the eigenvalue decomposition of        


##### then we notice that by pre-multiplying both sides of the equation with T, we obtain

                                         
##### Meaning that, if ui is an eigenvector of T T', then vi = Tui is an eigenvector of S. If we have a training set of 300 images of 64 × 64 pixels, the 
              
##### matrix T T' is a 300 × 300 matrix, which is much more manageable than the 4096 × 4096 covariance matrix. The obtained vector is feature scaled.

```python
def show_eigenfaces(pca):
	fig, axes = plt.subplots(3, 8, figsize=(9, 4),
	                         subplot_kw={'xticks':[], 'yticks':[]})
	for i, ax in enumerate(axes.flat):
	    ax.imshow(pca.components_[i].reshape(64, 64), cmap='gray')
	    ax.set_title("PC " + str(i+1))
	plt.show()
```

```python
print("printing eigen faces")
show_eigenfaces(pca)
```

```python
Xtrain_pca = pca.transform(x_train)
clf = SVC(kernel='rbf',C=1000,gamma=0.001)
clf = clf.fit(Xtrain_pca, y_train)
t0 = time()
Xtest_pca = pca.transform(x_test)
y_pred = clf.predict(Xtest_pca)
```

## Applying Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=clf,X= Xtrain_pca,y=y_train,cv=5,n_jobs=-1)
print (accuracies)
print("The mean of all the accuraices is")
print(accuracies.mean())
```

```python
import seaborn as sns
from sklearn import metrics
plt.figure(1, figsize=(12,8))
sns.heatmap(metrics.confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

## More Results
#### The ML model is subject to 4 other clustering algorithms 
#### 1. XGBOOST
#### 2. Gaussian Naive Bayes 
#### 3. Linear Discriminant Analysis
#### 4. k-nearest neighbors algorithm
#### 5. Decision tree learning

```python
models=[]
models.append(('XGBOOST', XGBClassifier()))
models.append(("LDR",LinearDiscriminantAnalysis()))
models.append(("NB",GaussianNB()))
models.append(("KNN",KNeighborsClassifier(n_neighbors=5)))
models.append(("DT",DecisionTreeClassifier()))


for name, model in models:
    
    clf=model

    clf.fit(Xtrain_pca, y_train)

    y_pred=clf.predict(Xtest_pca)
    print(10*"-","{} Result".format(name).upper(),10*"-")
    print("Accuracy score:{:0.2f}".format(accuracy_score(y_test, y_pred)))
    print()
```
