### The dataset is transformed into numbers. now let us read and visualize the data

```python
def show_images(img,val1,val2,fval1,fval2):
	fig, axes = plt.subplots(val1, val2, figsize=(fval1, fval2),
	                         subplot_kw={'xticks':[], 'yticks':[]})
	for i, ax in enumerate(axes.flat):
	    ax.imshow(np.array(img)[i].reshape(64, 64), cmap='gray')	
	plt.show()    
```

### Split Dataset into train and test set. we have set the test_size to 0.2 implies that take 20% of data for testing and train the ML model with 80% of the dataset. 

######  We have set the random state to 20 in this model. Random state ensures that the splits that you generate are reproducible. Scikit-learn uses random permutations to generate the splits. The random state that you provide is used as a seed to the random number generator. This ensures that the random numbers are generated in the same order. 

```python
sequential_face_target = pd.Series(faces_target)
x_train, x_test, y_train, y_test = train_test_split(face_data, sequential_face_target,test_size=0.2,random_state=20)
```

# Principal Component Analysis
### PCA is a machine learning model which is a part of unsupervised learning.Unsupervised learning is a type of machine learning algorithm used to draw inferences from datasets consisting of input data without labeled responses.

```python
pca = PCA(n_components=120).fit(x_train)
print(" printing the varience of each of the 120 components chosen \n")
print(pca.explained_variance_ratio_)
print('\n printing graphical representation')
plt.plot(np.cumsum(pca.explained_variance_ratio_),'r')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.show()
print('From the graph it can be seen that the curve nearly flattens after 120 components which indicates that out of all the 64*64 pixels (4096) features, 120 components cause the most variance')
```

#### The set of eigen vectors helps in locating the features that cause most varience hence intensively used in facial recognition

```python
Xtrain_pca = pca.transform(x_train)
Xtest_pca = pca.transform(x_test)
```

## Finding the best Algorithm
#### The ML model is subject to 4 other clustering algorithms 
#### 1. XGBOOST
#### 2. Gaussian Naive Bayes 
#### 3. Linear Discriminant Analysis
#### 4. k-nearest neighbors algorithm
#### 5. Decision tree learning

```python
models=[]
models.append(('XGBOOST', XGBClassifier()))
models.append(("LDA",LinearDiscriminantAnalysis(solver = 'eigen',shrinkage = 'auto')))
models.append(("NB",GaussianNB()))
models.append(("KNN",KNeighborsClassifier(n_neighbors=15)))
models.append(("DT",DecisionTreeClassifier()))

for name, model in models:
    
    clf=model

    clf.fit(Xtrain_pca, y_train)

    y_pred=clf.predict(Xtest_pca)
    print(10*"-","{} Result".format(name).upper(),10*"-")
    print("Accuracy score",(accuracy_score(y_test, y_pred)))
    print()
```

```python
print('from the above we see best results with LDA so we will use the LDA classifier')
clf = LinearDiscriminantAnalysis(solver = 'eigen',shrinkage = 'auto')
clf = clf.fit(Xtrain_pca, y_train)
y_pred=clf.predict(Xtest_pca)
print(accuracy_score(y_test,y_pred))
```

## Applying Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
accuracies = cross_val_score(estimator=clf,X= Xtrain_pca,y=y_train,cv=5,n_jobs=-1)
kfold = KFold(n_splits=5)
pred = cross_val_score(clf, Xtrain_pca, y_train, cv=kfold, scoring='accuracy')
print(pred)
print("The mean of all the accuraices is")
print(pred.mean())
```

```python
import seaborn as sns
from sklearn import metrics
plt.figure(1, figsize=(12,8))
sns.heatmap(metrics.confusion_matrix(y_test, y_pred))
```
